```python
def kClosest(points, k):
        """
        :type points: List[List[int]]
        :type k: int
        :rtype: List[List[int]]
        """
        distances = []
        for point in points:
            x = point[0]
            y = point[1]
            d = (x**2 + y**2)**0.5
            distances.append((d, point))
        distances.sort()
        print(distances)
        result = []
        for i in range(k):
            result.append(distances[i][1])
        return result
points = [[1,3],[-2,2]]
k = 1
kClosest(points, k)
```

```python
nums = [2,5,6,9,10]
def f(nums):
  min_num = min(nums)
  max_num = max(nums)
  result = 0
  for i in range(1, min(min_num, max_num)+1):
    if min_num % i ==0 and max_num % i == 0 :
      result = i
  return result
f(nums)
# gcd(5,10)
```

```python
def largest_perimeter(nums):
    nums.sort(reverse=True)   # sort descending

    for i in range(len(nums) - 2):
        a = nums[i]           # largest side
        b = nums[i + 1]
        c = nums[i + 2]       # smallest side

        if b + c > a:         # triangle condition
            return a + b + c  # first valid = largest perimeter

    return 0
k = [3,2,3,4]
largest_perimeter(k)
```

```python
#dynamic programming
def fib(n):
  d = {1:1,
       2:2}
  for i in range(3, n+1):
    d[i] = d[i-1] + d[i-2]
  return d[n]
fib(5)
#instead
# def fib2(n):
#   if n <= 2:
#     return n
#   a,b = 1,2
#   for _ in range(3, n+1):
#     a,b = b, a+b
#   return b
# fib2(6)
```

```python
#min cost climbing stairs
def climbing(cost):
  d = {0: cost[0],
       1: cost[1]}
  for i in range(2, len(cost)):
    d[i] =cost[i] + min(d[i-1], d[i-2])
  return min(d[len(cost)-1],d[len(cost)-2])
cost = [1,100,1,1,1,100,1,1,100,1]
climbing(cost)
```
