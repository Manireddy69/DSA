```python
class Node:
  def __init__(self, data):
    self.data = data
    self. next = None
class Linkedlist:
  def __init__(self):
    self.head = None
  def print_LL(self):
    curr = self.head
    while curr is not None:
      print(curr.data, end = '->')
      curr = curr.next
  def insert_front(self, num):
    curr = self.head
    self.head = Node(num)
    self.head.next = curr
  def insert_last(self, num):
    curr = self.head
    while curr.next is not None:
      curr = curr.next
    curr.next = Node(num)
  def delete_head(self):
    self.head = self.head.next
    self.print_LL()
  def remove(self,num):
    curr = self.head
    while curr.next is not None:
      if curr.next.data == num:
        curr.next = curr.next.next
        return
      curr = curr.next
ll = Linkedlist()
ll.head = Node(8)
ll.head.next = Node(7)
ll.head.next.next = Node(0)
ll.delete_head()
```
