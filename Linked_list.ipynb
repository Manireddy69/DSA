```python
class Node:
  def __init__(self, data):
    self.data = data
    self. next = None
class Linkedlist:
  def __init__(self):
    self.head = None
  def print_LL(self):
    curr = self.head
    while curr is not None:
      print(curr.data, end = '->')
      curr = curr.next
  def insert_front(self, num):
    curr = self.head
    self.head = Node(num)
    self.head.next = curr
  def insert_last(self, num):
    curr = self.head
    while curr.next is not None:
      curr = curr.next
    curr.next = Node(num)
  def delete_head(self):
    self.head = self.head.next
    self.print_LL()
  def remove(self,num):
    curr = self.head
    while curr.next is not None:
      if curr.next.data == num:
        curr.next = curr.next.next
        return
      curr = curr.next
  def count(self):
    curr = self.head
    count = 0
    while curr is not None:
      count += 1
      curr = curr.next
    return count
  def search(self, num):
    curr = self.head
    while curr is not None:
      if curr.data == num:
        return True
      curr = curr.next
    return False
```

```python
ll = Linkedlist()
ll.head = Node(8)
ll.head.next = Node(7)
ll.head.next.next = Node(0)
ll.count()
print(ll.print_LL())
print(ll.search(1))
```

```python
# Redefining Node and Doublylinkedlist to correct issues
class Node:
  def __init__(self,data): # Corrected from __int__ to __init__
    self.data = data
    self.prev = None
    self.next = None

class Doublylinkedlist:
  def __init__(self):
    self.head = None
    self.tail = None # Make append O(1)
  def print(self):
    curr = self.head # Corrected: should use self.head, not a global 'head'
    while curr is not None:
      print(curr.data, end = '<->')
      curr = curr.next
  #assigning this Linked list manually is too tedious
  #so we use append method and assing the pointers
  def append(self, num):
    new_node = Node(num)
    if not self.head:
      self.head = new_node
      self.tail = new_node
    else:
      new_node.prev = self.tail
      self.tail.next = new_node
      self.tail = new_node
  def length(self): # same as singly linked list
    curr = self.head
    count = 0
    while curr is not None:
      count += 1
      curr = curr.next
    return count
```
