```python
def fib(n, memo = {}):
  if n in memo :
    return memo[n]
  if n<= 1:
    return n
  memo[n] = fib(n-1, memo) + fib(n-2, memo)
  return memo[n]

fib(5) # 5th term of fibnacci series, which is 5(0,1,1,2,3,5)
```

```python
#bottom up approach
def fib_s(n):
  if n<= 2:
    return 1
  fib_s = [0] * (n+1) #it set's (n+1) into 0's
  fib_s[1], fib_s[2]  = 1,1
  for i in range(3, n+1):
    fib_s[i] = fib_s[i-1] + fib_s[i-2]
  return fib_s[n]
fib_s(5)
```

```python
#knapsack problem
#brute force or recursive method
weights = [1,3,4,5]
val = [1,4,5,7]
capacity = 7
def knapsack_recursive(weights,val,capacity, n):
  if n == 0 or capacity == 0:
    return 0
  if weights[n-1] > capacity:
    return knapsack_recursive(weights,val, capacity, n-1)
  else:
    return max(val[n-1]+knapsack_recursive(weights,val, capacity - weights[n-1], n-1), knapsack_recursive(weights,val, capacity, n-1))
# knapsack_recursive(weights, val, capacity, len(weights))
def knapsack_memo(weights, val,capacity):
  n = len(weights)
  dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]
  for i in range(1, n+1):
    for w in range(1, capacity + 1):
      if weights[i-1] <= w:
        dp[i][w] = max(val[i-1] + dp[i-1][w-weights[i-1]], dp[i-1][w])
      else:
        dp[i][w] = dp[i-1][w]
  print(dp)
  return dp[n][capacity]
knapsack_memo(weights, val, capacity)
```

```python
arr1 = [1,2,3,4,5]
arr2 = [3,4,5,1,2]

def f(arr1, arr2):
  k = 0
  def left(arr1, k):
    if arr1 == arr2:
      return k
    else:
      return left(arr1[1:]+arr1[:1], k + 1)
  l = 0
  def right(arr1, l):
    if arr1 == arr2:
      return l
    else:
      return right(arr1[-1:]+arr1[:-1], l + 1)
  return min(left(arr1, k), right(arr1, l))
f(arr1, arr2)
```

```python
nums = [2,3,-1,4]
def f(nums):
  min_prod = nums[0]
  max_prod = nums[0]
  result = nums[0]
  for i in nums[1:]:
    candidates = (i, max_prod*i , min_prod*i)
    max_prod = max(candidates)
    min_prod = min(candidates)
    result = max(result, max_prod)
  return result
f(nums)
```

```python
def is_prime(num):
  if num < 2:
    return False
  for i in range(2, int(num**0.5) + 1):
    if num % i == 0:
      return False
  return True

def find_first_n_primes(n):
  primes = []
  if n <= 0:
    return primes

  # We need to iterate over a range of numbers until we find 'n' primes.
  # A generous upper bound is used to ensure 'n' primes are found.
  # For example, to find the 100th prime, we might need to check numbers up to around 600.
  # A heuristic of n * 100 should be sufficient for typical values of n.
  # We start checking from 2.
  for num in range(2, max(100, n * 100)): # Ensures a sufficiently large range
    if len(primes) == n: # Stop once 'n' primes are found
      break
    if is_prime(num):
      primes.append(num)
  return primes

# Example usage:
# print(find_first_n_primes(5)) # Outputs the first 5 prime numbers
# print(find_first_n_primes(10)) # Outputs the first 10 prime numbers
```

```python
nums = [1,5,11,5]
def f(nums):
  total = sum(nums)
  if total % 2 != 0:
    return False
  target = total //2
  dp = [False] * (target + 1)
  dp[0] = True
  for num in nums:
    for j in range(target, num-1, -1):
      dp[j] = dp[j] or dp[j-num]
  return dp[target]
f(nums)
```

```python
def knapsack(weights, values, capacity):
  n = len(weights)
  dp = [[0 for _ in range(capacity +1)]for _ in range(n+1)]  # create a zeros matrix with capacity +1 rows and n+1 cols

  for i in range(1, n+1):
    for w in range(1, capacity +1 ):
      if weights[i-1] <= w:
        dp[i][w] = max(values[i-1] + dp[i-1][w-weights[i-1]], dp[i-1][w])
      else:
        dp[i][w] = dp[i-1][w]
  print(dp)
  return dp[n][capacity]


weights = [1,2,3,4]
values = [100, 300, 350,500]
capacity = 5
print("manximum value in knapsack:", knapsack(weights, values, capacity))
```

```python
#dynamic programming
dic = {0 : 0,
       1 : 1,
       2 : 1}
def fib(n, dic = dic):
  if n in dic:
    return dic[n]
  else:
    return fib(n-1, dic) + fib(n-2, dic)

fib(4)
```

```python
#tabbulation method
def fib_s(n):

  if n == 0:
    return 0
  elif n == 1:
    return 1
  fib = [0] * (n+1)
  fib[1] = 1
  for i in range(2, n+1):
    fib[i] = fib[i-1] + fib[i-2]
  return fib[n]

fib_s(4)
```

```python
def binomial(n,c,sum = {}):  #tuples takes only number
  if c == 0 or n ==c:
    return 1
  if (n,c) in sum:
    return sum[(n,c)]
  sum[(n,c)] = binomial(n-1, c-1, sum) + binomial(n-1, c) #ncr = (n-1)c(r-1) +nc(r)
  return sum[(n,c)]

binomial(5,2)
```
