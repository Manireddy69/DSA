```python
matrix = [[1,2,3],[4,5,6],[7,8,9]]
def spiral(matrix):
  left = 0
  right= len(matrix[0])-1
  top = 0
  bottom  = len(matrix)-1
  result = []
  while top <= bottom and left <= right:
    for i in range(left, right +1 ):
      result.append(matrix[top][i])
    top +=1
    for j in range(top, bottom+1):
      result.append(matrix[j][right])
    right -= 1
    if top  <= bottom:
      for c in range(right, left-1, -1):
        result.append(matrix[bottom][c])
    bottom -= 1
    if left <= right:
      for r in range(bottom, top -1, -1):
        result.append(matrix[r][left])
      left += 1
  return result
spiral(matrix)
```

```python
def binary_search(nums, target):
  l = 0
  h = len(nums) - 1

  while l <= h:
    mid = (l + h) // 2

    if nums[mid] == target:
      return mid

    # Determine which half is sorted
    if nums[l] <= nums[mid]:  # Left half is sorted
      if nums[l] <= target < nums[mid]:  # Target is in the sorted left half
        h = mid - 1
      else:  # Target is in the unsorted right half
        l = mid + 1
    else:  # Right half is sorted
      if nums[mid] < target <= nums[h]:  # Target is in the sorted right half
        l = mid + 1
      else:  # Target is in the unsorted left half
        h = mid - 1
  return -1

# Test cases
nums1 = [4, 5, 6, 7, 0, 1, 2]
target1 = 0
print(f"Searching for {target1} in {nums1}: Index = {binary_search(nums1, target1)}") # Expected: 4

nums2 = [4, 5, 6, 7, 0, 1, 2]
target2 = 3
print(f"Searching for {target2} in {nums2}: Index = {binary_search(nums2, target2)}") # Expected: -1

nums3 = [1]
target3 = 1
print(f"Searching for {target3} in {nums3}: Index = {binary_search(nums3, target3)}") # Expected: 0

nums4 = [1, 3]
target4 = 3
print(f"Searching for {target4} in {nums4}: Index = {binary_search(nums4, target4)}") # Expected: 1

nums5 = [5, 1, 3]
target5 = 5
print(f"Searching for {target5} in {nums5}: Index = {binary_search(nums5, target5)}") # Expected: 0

nums6 = [3, 1]
target6 = 1
print(f"Searching for {target6} in {nums6}: Index = {binary_search(nums6, target6)}") # Expected: 1
```

```python
def  armstrong(n):
  l = len(str(n))
  num = n
  sum = 0
  while n >0:
    lk = n %10
    sum += lk ** l
    n = n//10
  return sum == num

armstrong(153)
```

```python
def fib(n):
  d = {0:0,
       1:1,
       2:1,
       3:2}
  if n in d:
    return d[n]
  d[n] = fib(n-1) + fib(n-2)
  return d[n]
fib(5)
```

```python
class Solution(object):
    def fib(self, n):
        """
        :type n: int
        :rtype: int
        """
        d = {0:0,1:1,2:1,3:2}
        if n in d:
            return d[n]
        d[n] = self.fib(n-1) + self.fib(n-2)
        return d[n]

sol = Solution()
f = sol.fib(5)
print(f)
```

```python
class Solution(object):
    def __init__(self):
        self.memo = {0: 0, 1: 1, 2: 1}

    def tribonacci(self, n):
        """
        :type n: int
        :rtype: int
        """
        if n in self.memo:
            return self.memo[n]

        # Calculate and store the result
        self.memo[n] = self.tribonacci(n - 1) + self.tribonacci(n - 2) + self.tribonacci(n - 3)
        return self.memo[n]

sol = Solution()
f = sol.tribonacci(5)
print(f)
```

```python
#SORTING
n = [7,4,1,5,3]
def selection(n):
  l = len(n)
  for i in range(l-1):
    min = i
    for j in range(i+1,l):
      if n[min] > n[j]:
        min = j
      # print("The elements after swap",j,n)
    # print("The elements after swap",i+1,n)
    n[i], n[min] = n[min], n[i]
  return n
selection(n)
```

```python
def f(n):
  stack = []

  for char in n:
    if char == '(' or char == '[' or char == '{': # If it's an opening bracket
      stack.append(char)
    elif char == ')': # If it's a closing parenthesis
      if not stack or stack.pop() != '(': # Check if stack is empty or top doesn't match
        return False
    elif char == ']': # If it's a closing square bracket
      if not stack or stack.pop() != '[': # Check if stack is empty or top doesn't match
        return False
    elif char == '}': # If it's a closing curly brace
      if not stack or stack.pop() != '{': # Check if stack is empty or top doesn't match
        return False
    # Optionally, handle non-bracket characters here, currently they are ignored.

  # If the stack is empty, all brackets were matched
  return not stack

# Test cases
n1 = "[()]{}"
print(f"'{n1}': {f(n1)}") # Expected: True

n2 = "([{}])"
print(f"'{n2}': {f(n2)}") # Expected: True

n3 = "([)]"
print(f"'{n3}': {f(n3)}") # Expected: False

n4 = "{"
print(f"'{n4}': {f(n4)}") # Expected: False

n5 = "}"
print(f"'{n5}': {f(n5)}") # Expected: False

n6 = "" # Empty string
print(f"'{n6}': {f(n6)}") # Expected: True

n7 = "{[]()}"
print(f"'{n7}': {f(n7)}") # Expected: True

n8 = "((("
print(f"'{n8}': {f(n8)}") # Expected: False

n9 = ")))"
print(f"'{n9}': {f(n9)}") # Expected: False
```
